```python
#2
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
a, b, c = 1, -2, 3
y = a * x**2 + b * x + c

plt.figure()
plt.plot(x, y, label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График квадратичной функции')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 400)
A, B, C = 1, 2, np.pi/4
y = A * np.sin(B * x + C)

plt.figure()
plt.plot(x, y, label=f'y = {A}*sin({B}x + {C:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График гармонической функции')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt

x_quad = np.linspace(-10, 10, 400)
a, b, c = 1, -2, 3
y_quad = a * x_quad**2 + b * x_quad + c

x_harm = np.linspace(0, 10, 400)
A, B, C = 1, 2, np.pi/4
y_harm = A * np.sin(B * x_harm + C)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(x_quad, y_quad, label=f'y = {a}x^2 + {b}x + {c}')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График квадратичной функции')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x_harm, y_harm, label=f'y = {A}*sin({B}x + {C:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График гармонической функции')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#5
import numpy as np
import matplotlib.pyplot as plt

x_quad = np.linspace(-10, 10, 400)
a, b, c = 1, -2, 3
y_quad = a * x_quad**2 + b * x_quad + c

x_harm = np.linspace(0, 10, 400)
A, B, C = 1, 2, np.pi/4
y_harm = A * np.sin(B * x_harm + C)

plt.figure(figsize=(8, 6))

plt.plot(x_quad, y_quad, label=f'y = {a}x^2 + {b}x + {c}', color='blue')

plt.plot(x_harm, y_harm, label=f'y = {A}*sin({B}x + {C:.2f})', color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Графики квадратичной и гармонической функций')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#6
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4 * np.pi, 400)
x = np.cos(3 * t)
y = np.sin(2 * t)

plt.figure()
plt.plot(x, y, label='x(t) = cos(3t), y(t) = sin(2t)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('График функции, заданной параметрически')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#7
import numpy as np
import matplotlib.pyplot as plt

years = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
values = np.array([4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32])

plt.figure()
plt.bar(years, values, color='skyblue')
plt.xlabel('Год')
plt.ylabel('Прожиточный минимум')
plt.title('Прожиточный минимум в Кыргызской Республике (2012-2022)')
plt.xticks(years)
plt.grid(axis='y')
plt.show()
```

```python
#8
import matplotlib.pyplot as plt

categories = ['2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']
values = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.figure()
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(categories))))
plt.axis('equal')  # Отображение диаграммы как круга, а не эллипса
plt.title('Доля прожиточного минимума по годам')
plt.show()
```
